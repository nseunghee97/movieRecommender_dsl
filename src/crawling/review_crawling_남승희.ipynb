# 네이버 영화 리뷰 크롤링

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
movie_codes = {
    '아이 캔 스피크' : ['161850','2017','드라마'],
    '덕혜옹주' : ['94767','2016','서사,드라마'] ,
    '신과함께2' : ['167697','2017','판타지,드라마'],
    '1987' : ['158191','2017','드라마'],
    '터널' : ['141104','2016','드라마'],
    '타짜-신의 손' : ['67769','2014','드라마'],
    '아저씨' : ['71509','2010','액션,범죄,드라마'],
    '검사외전' : ['130903','2016','범죄,코미디'],
    '변호인' : ['101901','2013','드라마'],
    '택시운전사' : ['146469','2017','드라마,가족'],
    '럭키' : ['140695','2015','코미디']
}
```

```python
list(movie_codes.values())[0][1]
```

```python
title = []
year = []
genre = []
score = []
review = []

for i in range(len(movie_codes)):
    movie_title = list(movie_codes.keys())[i]
    movie_code = list(movie_codes.values())[i][0]
    movie_year = list(movie_codes.values())[i][1]
    movie_genre = list(movie_codes.values())[i][2]
    for j in range(100):
        url = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=' + movie_code + '&type=after&onlyActualPointYn=Y&onlySpoilerPointYn=N&order=sympathyScore&page=' + str(j+1)
        res = requests.get(url)
        soup = BeautifulSoup(res.text,'html.parser')
        for r in soup.select('div.score_reple > p > span:nth-child(2)'):
            r = r.text.strip()
            review.append(r)
            title.append(movie_title)
            genre.append(movie_genre)
            year.append(movie_year)
        for s in soup.select('div.star_score > em'):
            score.append(s.text.strip())
```

```python
len(title),len(year),len(genre),len(score),len(review)
```

```python
movie_review = {
    'title': title,
    'year' : year,
    'genre': genre,
    'score' : score,
    'review': review
}  
    
review_df = pd.DataFrame(movie_review)
review_df
```

```python
review_df.to_csv('review_naver_남승희.csv',index = False)
```
